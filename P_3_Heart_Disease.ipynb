##displot (Distribution Plot):

displot is primarily used for numerical variables to show the distribution of data points. It helps you understand the frequency or density of different values within a continuous or discrete numerical range (e.g., age, cholesterol levels, blood pressure). It often includes histograms and Kernel Density Estimates (KDEs).
Example from your notebook: You used sns.displot(df['Age'], kde=True) which is a perfect use case as 'Age' is a numerical variable.
##Pie Chart:

Pie charts are best suited for categorical variables to display the proportion of each category relative to the whole. They are most effective when you have a small number of categories (ideally 2 to 5) to avoid clutter and make it easy to compare segments.
Example from your notebook: You used pie charts for df['Sex'] and df['ChestPainType'], which are categorical variables, to show their distribution in percentages. ST_Slope is also a good example.
In summary, use displot for understanding the shape and spread of numerical data, and use pie charts for visualizing the proportional breakdown of categorical data.

### Age Distribution by Heart Disease Status (Histogram)

This histogram will show the frequency distribution of 'Age' for individuals, separated by whether they have 'HeartDisease' or not. This helps in understanding if age is a discriminating factor.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='HeartDisease', multiple='stack', kde=True)
plt.title('Age Distribution by Heart Disease Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='HeartDisease', labels=['No', 'Yes'])
plt.show()
```

### Age Distribution by Heart Disease Status (Box Plot)

This box plot will provide a summary of the 'Age' distribution (median, quartiles, and outliers) for each 'HeartDisease' category, offering a clear comparison of age ranges.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='HeartDisease', y='Age')
plt.title('Age Distribution by Heart Disease Status')
plt.xlabel('Heart Disease (0: No, 1: Yes)')
plt.ylabel('Age')
plt.show()
```

```python
df.groupby('HeartDisease').size().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title("Heart Disease Distribution")
plt.ylabel("")   # removes unnecessary ylabel
```

### Correlation Heatmap

A correlation heatmap visualizes the pairwise correlation coefficients between all numerical variables in the DataFrame. The color intensity and legend indicate the strength and direction (positive or negative) of the correlation. This helps in identifying highly correlated features.

Certainly! I'll create a correlation heatmap to visualize the relationships between the numerical variables in your dataset. This plot will help us understand how strongly each pair of numerical features is correlated with each other.

The correlation heatmap has been successfully generated. This visualization is very useful for understanding the relationships between the numerical variables in your dataset.

#Color Scale:
 The cmap='coolwarm' colormap indicates positive correlations with warmer colors (reds) and negative correlations with cooler colors (blues). A value close to 1 means a strong positive correlation, -1 means a strong negative correlation, and 0 means no linear correlation.
annot=True and fmt=".2f": These parameters display the actual correlation coefficients on each cell, formatted to two decimal places, making it easy to read the exact strength and direction of the relationship.
##How to interpret:

By looking at the heatmap, you can quickly identify which features are highly correlated (either positively or negatively). For example, if you see a bright red square between 'Age' and 'HeartDisease', it suggests a strong positive correlation, meaning as age increases, the likelihood of heart disease also tends to increase. Conversely, a strong blue square would indicate a negative correlation.
Pay attention to correlations with the 'HeartDisease' column. These values can give us initial insights into which features might be strong predictors for heart disease.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
