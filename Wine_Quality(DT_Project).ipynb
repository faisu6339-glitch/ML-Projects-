## Scale Numerical Features and Train Logistic Regression

### Subtask:
Identify numerical features, apply StandardScaler to numerical features in both X_train and X_test, and then initialize and train a Logistic Regression model with specified parameters on the scaled data.


**Reasoning**:
The subtask requires identifying numerical features, scaling them using `StandardScaler`, and then training a `LogisticRegression` model on the scaled data. This step will perform all these actions in a single code block, ensuring all numerical features are scaled correctly before training the model, and setting the `random_state` for reproducibility.



```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import warnings

# 1. Identify numerical features (all columns except 'type_white')
numerical_features = X_train.columns.drop('type_white').tolist()

# 2. and 3. Instantiate StandardScaler
scaler = StandardScaler()

# 4. Apply fit_transform to numerical features of X_train
X_train[numerical_features] = scaler.fit_transform(X_train[numerical_features])

# 5. Apply transform to numerical features of X_test
X_test[numerical_features] = scaler.transform(X_test[numerical_features])

# Suppress the FutureWarning related to multi_class (if it appears)
with warnings.catch_warnings():
    warnings.simplefilter("ignore", FutureWarning)
    # 6. and 7. Instantiate Logistic Regression model with specified parameters
    log_model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=5000, random_state=42)

    # 8. Train the model
    log_model.fit(X_train, y_train)

print("Numerical features scaled and Logistic Regression model trained successfully.")
```

## Make Predictions with Logistic Regression

```python
y_pred_log_reg = log_model.predict(X_test)
print("Predictions made successfully using the Logistic Regression model.")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# Hyperparameter grid
grid_param = {
    'n_estimators': [5, 10, 50, 100, 120, 150],
    'criterion': ['gini', 'entropy'],
    'max_depth': list(range(2, 32)),
    'min_samples_leaf': list(range(1, 10))
}

# Random Forest with class imbalance handling
rf = RandomForestClassifier(
    random_state=42,
    class_weight='balanced',
    n_jobs=-1
)

# Grid Search
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=grid_param,
    cv=5,
    scoring='f1_macro',   # BEST for imbalanced multiclass
    verbose=1,
    n_jobs=-1
)

# Train
grid_search.fit(X_train, y_train)

# Best model
best_rf = grid_search.best_estimator_

print("Best Parameters:")
print(grid_search.best_params_)

# Predictions
y_pred = best_rf.predict(X_test)

# Evaluation
print("\nRandom Forest Accuracy:",
      accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
