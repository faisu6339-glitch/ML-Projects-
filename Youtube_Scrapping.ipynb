```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(browser.page_source, "html.parser")
sp = soup.find_all("ytd-grid-video-renderer")

titles = []
video_links = []
views = []
dates = []

for item in sp:

    # Title & Link
    video_title = item.find("a", class_="yt-simple-endpoint style-scope ytd-grid-video-renderer")

    if video_title:
        titles.append(video_title.text.strip())
        video_links.append(video_title.get("href"))

    # Views & Date
    span_data = item.find_all("span", class_="style-scope ytd-grid-video-renderer")

    # Avoid index error
    if len(span_data) >= 2:
        views.append(span_data[0].text.strip())
        dates.append(span_data[1].text.strip())
    else:
        views.append("")
        dates.append("")

# Display results
for i in range(len(titles)):
    print(f"{i+1}. Title: {titles[i]}")
    print(f"   Link : https://www.youtube.com{video_links[i]}")
    print(f"   Views: {views[i]}")
    print(f"   Date : {dates[i]}\n")

print("Would you like to save this data to a pandas DataFrame or CSV file?")
```

Thumbnail Extraction: It attempts to find an <img> tag within the video item and extracts the thumbnail URL from its src, data-thumb, or data-src attributes.

```python
import requests
from bs4 import BeautifulSoup

data = []   # <-- All results stored here

for sp in soup.find_all('ytd-grid-video-renderer'):

    # ---- Title & Link ----
    try:
        a_tag = sp.find('a', class_='yt-simple-endpoint style-scope ytd-grid-video-renderer')
        title = a_tag.text.strip()
        video_link = a_tag.get('href')
        full_link = "https://www.youtube.com" + video_link
    except:
        continue

    # ---- Views ----
    try:
        views = sp.find_all('span', class_='style-scope ytd-grid-video-renderer')[0].text.strip()
    except:
        views = None

    # ---- Date ----
    try:
        date_time = sp.find_all('span', class_='style-scope ytd-grid-video-renderer')[1].text.strip()
    except:
        date_time = None

    # ---- Thumbnail (Simple & Reliable) ----
    try:
        img = sp.find("img")
        thumbnail_link = (
            img.get("src")
            or img.get("data-src")
            or img.get("data-thumb")
            or img.get("srcset")
        )

        # If srcset → take first URL
        if thumbnail_link and "," in thumbnail_link:
            thumbnail_link = thumbnail_link.split(",")[0].strip().split(" ")[0]

        # Fallback thumbnail
        if not thumbnail_link or thumbnail_link.startswith("data:image"):
            if "v=" in video_link:
                video_id = video_link.split("v=")[1]
            else:
                video_id = video_link.split("/")[-1]

            thumbnail_link = f"https://i.ytimg.com/vi/{video_id}/hqdefault.jpg"

    except:
        thumbnail_link = None

    # ---- Description (from video page) ----
    try:
        page = requests.get(full_link, headers={"User-Agent":"Mozilla/5.0"}).text
        v_soup = BeautifulSoup(page, "html.parser")
        meta_desc = v_soup.find("meta", {"name": "description"})
        description = meta_desc.get("content") if meta_desc else ""
    except:
        description = ""

    # ---- STORE IN data VARIABLE ----
    data.append([
        title,
        full_link,
        views,
        date_time,
        thumbnail_link,
        description
    ])


```

##GFG Code To Access the videos

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

data = []

# Your channel videos page
url = "https://www.youtube.com/@geeksforgeeks/videos"

html = requests.get(url, headers={"User-Agent": "Mozilla/5.0"}).text
soup = BeautifulSoup(html, "html.parser")

for sp in soup.find_all('ytd-grid-video-renderer'):

    # ----- TITLE & FULL LINK -----
    try:
        a_tag = sp.find('a', class_='yt-simple-endpoint style-scope ytd-grid-video-renderer')
        title = a_tag.text.strip()
        video_link = a_tag.get("href")
        full_link = "https://www.youtube.com" + video_link
    except:
        title = ""
        full_link = ""

    # ----- VIEWS -----
    try:
        views = sp.find_all('span', class_='style-scope ytd-grid-video-renderer')[0].text.strip()
    except:
        views = ""

    # ----- DATE -----
    try:
        date_time = sp.find_all('span', class_='style-scope ytd-grid-video-renderer')[1].text.strip()
    except:
        date_time = ""

    # ----- THUMBNAIL -----
    try:
        img = sp.find("img")
        thumbnail = (
            img.get("src")
            or img.get("data-src")
            or img.get("data-thumb")
            or img.get("srcset")
        )
        # srcset → pick first URL
        if thumbnail and "," in thumbnail:
            thumbnail = thumbnail.split(",")[0].strip().split(" ")[0]

        # fallback thumbnail
        if not thumbnail or thumbnail.startswith("data:image"):
            if "v=" in video_link:
                vid = video_link.split("v=")[1]
            else:
                vid = video_link.split("/")[-1]
            thumbnail = f"https://i.ytimg.com/vi/{vid}/hqdefault.jpg"

    except:
        thumbnail = ""

    # ----- DESCRIPTION (From video page) -----
    try:
        page = requests.get(full_link, headers={"User-Agent": "Mozilla/5.0"}).text
        v_soup = BeautifulSoup(page, "html.parser")
        desc_meta = v_soup.find("meta", {"name": "description"})
        description = desc_meta.get("content") if desc_meta else ""
    except:
        description = ""

    # Append in the SAME style as your example
    data.append([title, views, date_time, thumbnail, description, full_link])

```
